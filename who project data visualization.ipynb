# Data Visualization

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\uzing\\Documents\\who_mortality_data.csv")
```

```python
country=df.Country
```

```python
percentage=df.Liver_disease
```

```python
labels=country
sizes=percentage
```

```python
plt.pie(sizes,labels=labels, autopct='%1.2f%%')
plt.axis("equal")
plt.legend(title="Countries",loc="upper right",bbox_to_anchor=(1.11,0,0.3,1))
plt.title("Liver Disease % Distribution")
plt.show()
```

```python
df2=pd.read_csv("C:\\Users\\uzing\\Documents\\asthma_data.csv")
```

```python
all_deaths=df2.All
```

```python
male=df2.Male
```

```python
female=df2.Female
```

```python
country1=df2.Country
```

```python
w=0.3
```

```python
alldeaths_bar=np.arange(len(country1))
females_bar=[i+w for i in alldeaths_bar]
males_bar=[i+w+w for i in alldeaths_bar]
plt.figure(figsize=(10,6))
plt.bar(alldeaths_bar,all_deaths,width=w,label="All deaths")
plt.bar(females_bar,female,width=w,label="Females")
plt.bar(males_bar,male,width=w,label="Males")
plt.xticks(alldeaths_bar+w,country1)
plt.title("Percentage Distrubution of Asthma Deaths by Gender")
plt.xlabel("Countries")
plt.ylabel("%  of deaths")
plt.legend()
plt.show()
plt.savefig('imagegraph.png')
plt.close()
```
